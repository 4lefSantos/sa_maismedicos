# Predictive analysis of Mais Médicos retention

Importing libraries

```python
import pandas as pd
from pycaret.classification import *
```

Reading dataframe

```python
df_original = pd.read_excel("base_tratadav2.xlsx")
df_original.head()
```

Selecting only features which will be useful 

```python
df = df_original[['uf_destino', 'sexo', 'anos_atuacao_mesmo_municipio', 'media_vinculos', 'churn', 'idade_em_anos', 
                  'anos_formacao', 'leitos_instalacao_fisica_urgencia', 'enfermeiro', 'tec_aux_enf',
                 'agente_saude', 'dentista', 'populacao', 'porte', 'pib_percapita', 'tx_pop_ocupada', 'cobertura_saude',
                 'vagas_medicina_milhab', 'orcamento_saude', 'esgotamento_sanitario', 'taxa_homicidio', 'ideb_anosiniciais',
                 'ideb_anosfinais', 'investimento_infraestrutura', 'equipamentos_esportivos', 'equpamentos_culturais',
                 'distancia_capital']]
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
setup_classificacao = setup(data = df, target = 'churn', train_size = 0.75, session_id = 2625, remove_multicollinearity=True,
                           multicollinearity_threshold = 0.7, ordinal_features = {"porte":["PP I", "PP II", "MP", "GP", "Metropole"]})
```

```python
help(setup)
```

```python
modelos = compare_models(n_select = 5)
```

```python
type(modelos[0])
```

```python
ridge_tunado = tune_model(modelos[0])
```

```python
modelo_final = finalize_model(ridge_tunado)
```

```python
modelo_final
```

```python
plot_model(modelo_final, plot='feature')
```
